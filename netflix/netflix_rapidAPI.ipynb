```python
def change_String(change_string):
    change_string = change_string.replace("}","")
    change_string = change_string.replace("{","")
    change_string = change_string.replace("\"","")
    change_string = change_string.replace("[","")
    change_string = change_string.replace("]","")
    return change_string
```

```python

def start_Crwl_Netflix(count):
    key = []
    value = []


    for i in range(count):
        response = crwl_Netflix(str(i))
        response2 = response.text.split(',')

        for j in range(len(response2)):
            dt = response2[j].split(":")
            if dt[0].find('COUNT') != -1:
                continue

            if len(dt) == 3:
                if j==1:
                    key.append(dt[1])
                    value.append(dt[2])
                else:
                    key.append(dt[0])
                    value.append(dt[1]+dt[2])

            elif len(dt) == 1:
                continue

            else:

                try:
                    key.append(dt[0])
                    value.append(dt[1])
                except IndexError:
                    break;


    if len(key) == len(value):
        for k in range(len(key)):
            key[k] = change_String(key[k])
            value[k] = change_String(value[k])
        dt_netflix2 = [ x for x in zip(key,value) ]

    else:
            print("개수가 달라요")

    dt_netflix = [ x for x in zip(key,value) ]
    return dt_netflix
```
