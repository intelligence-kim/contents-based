```python
from selenium import webdriver
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
import datetime
from itertools import count
import ssl 
import sys
import re      
import time
```

```python
def isHangul(text):
    #Check the Python Version
    pyVer3 =  sys.version_info >= (3, 0)

    if pyVer3 : # for Ver 3 or later
        encText = text
    else: # for Ver 2.x
        if type(text) is not unicode:
            encText = text.decode('utf-8')
        else:
            encText = text

    hanCount = len(re.findall(u'[\u3130-\u318F\uAC00-\uD7A3]+', encText))
    return hanCount > 0
```

```python
def list_Extraction(item):
    list_extraction = []
    try:
        l_item = list(item)
    except TypeError:
        return
        
    for i in range(len(l_item)):
        check = str(l_item[i]).find('genres')
        
        if isHangul(str(l_item[i])) == False:
            continue
            
        if check == -1 and l_item[i].string == None:
            list_extraction.append(str(l_item[i])[str(l_item[i]).find('>')+1 : str(l_item[i]).find('!')-1 ])
            
        elif check != -1:
            if str(l_item[i]).count('>') == 4:
                l_index = str(l_item[i])[:str(l_item[i])[:str(l_item[i]).rfind('>')].rfind('>')].rfind('>')
                r_index = str(l_item[i])[:str(l_item[i]).rfind('<')].rfind('<')

            else:
                l_index = str(l_item[i]).find('>')
                r_index = str(l_item[i])[:str(l_item[i]).rfind('<')].rfind('<')
            list_extraction.append(str(l_item[i])[l_index+1 : r_index ])
            
            
        else:
            list_extraction.append(l_item[i].string)
            
    if len(list_extraction) != 0:
        return list(set(list_extraction))
    
    else:
        return list(list_extraction)

```

```python
def check_Genres_Parameter(netflixid, imdbid, not_find):
    wd.get('https://www.netflix.com/browse/person/40009481?jbv='+str(netflixid))
    rcv_data = wd.page_source
    soupData = BeautifulSoup(rcv_data, 'html.parser')
    time.sleep(2)
    
    title = soupData.findAll('h2', attrs={'class':'section-subheader-text section-item'})
    if title != []:
        label = soupData.find('a', attrs={'class':'title-info-metadata-item item-genre'})
        program_character = soupData.findAll('span', attrs={'class':'more-details-item item-mood-tag'})
        genres = soupData.findAll('span', attrs={'class':'more-details-item item-genres'})
        cast = soupData.findAll('span', attrs={'class':'title-data-info-item-list'})
        season = soupData.findAll('span', attrs={'class':'duration'})

    else:
        title = soupData.findAll('h1', attrs={'class':'title-title'})
        if title == []:
            not_find.append(netflixid)
        label = soupData.find('span', attrs={'data-uia':'more-details-item-genres'})
        program_character = soupData.findAll('span', attrs={'class':'more-details-item item-mood-tag'})
        genres = soupData.findAll('span', attrs={'class':'more-details-item item-genres'})
        cast = soupData.findAll('span', attrs={'class':'title-data-info-item-list'})    
        season = soupData.findAll('span', attrs={'class':'duration'})

    return  {'netflixid' : str(netflixid),
             'title' : list_Extraction(title), 
             'label' : list_Extraction(label),
             'genres' : list_Extraction(genres), 
             'character' : list_Extraction(program_character),
             'cast' : list_Extraction(cast), 
             'season' : list_Extraction(season),
             'imdbid' : str(imdbid)}

```

```python
dict_list = []
not_find = []
for i in range(len(netflix_api_data)):
    if i % 100 == 10:
        print(i)
    official_data = check_Genres_Parameter(netflix_api_data.index[i],
                                            netflix_api_data.loc[netflix_api_data.index[i]]['imdbid'],
                                            not_find)
    dict_list.append(official_data)
    
```
