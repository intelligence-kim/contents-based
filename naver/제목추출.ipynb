```python
import json

with open('D:/Python/netflix_basic_data/netflix.json', encoding='UTF8') as json_file:
    json_data = json.load(json_file)
    f = open('D:/Python/netflix_basic_data/search_title.txt', 'w+', encoding='UTF8')
    f1 = open('D:/Python/netflix_basic_data/undefined_title.txt', 'w+', encoding='UTF8')
    title_list = ''
    undefined = ''

    count = 1
    for i in json_data:
        title_str = i["title"]
        netflixid_str = i["netflixid"]
        title=''.join(title_str)
        nt_id =''.join(netflixid_str)
        
        for i in new_line:
            if(nt_id == i):
                if(title == ''):
                    undefined += i + '\n'
                    continue
                print(count, " - " ,title)
                title_list += i + "\t" + title + '\n'
                count+=1

    f.write(title_list)
    f1.write(undefined)
    f.close()
    f1.close()
```
