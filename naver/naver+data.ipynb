```python
import pandas as pd
crwl_naver = pd.read_csv('result.csv', encoding='utf-8')
imdb = pd.read_csv('final_imdb.csv', encoding='utf-8')
```

```python
imdb
```

```python
crwl_naver
```

```python
crwl_naver = crwl_naver.rename({'netflix_id':'netflixid_y'}, axis='columns')
crwl_naver
```

```python
crwl_naver = crwl_naver.astype(str)
crwl_naver = crwl_naver.set_index('netflixid_y')
crwl_naver
```

```python
imdb = imdb.astype(str)
imdb = imdb.set_index('netflixid_y')
imdb
```

```python
merge_imdb_naver = pd.merge(imdb, crwl_naver, left_index=True, right_index=True, how='inner')
```

```python
len(merge_imdb_naver)
```

```python
merge_imdb_naver.to_csv('final_data.csv', encoding='utf-8')
```
