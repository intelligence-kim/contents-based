```python
import pandas as pd
import glob
import os
```

```python
input_file = r'D:\Python\netflix_basic_data\sum'
output_file = r'D:\Python\netflix_basic_data\sum\result.csv'
```

```python
allFile_list = glob.glob(os.path.join(input_file, 'total*.csv'))
print(allFile_list)
```

```python
allData = []
for file in allFile_list:
    df = pd.read_csv(file, names=['netflix_id','title','review1','review2','review3','review4','review5','review6','review7'], engine='python', encoding='utf-8-sig')
    allData.append(df)
```

```python
dataCombine = pd.concat(allData, axis=0, ignore_index=True)
dataCombine.to_csv(output_file, index=False, encoding='utf-8-sig')
```
