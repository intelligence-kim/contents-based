```python
import json
with open('/Users/studio/Desktop/netflix/netflix.json', 'r') as f:
    json_data = json.load(f)
print(json_data)
```

```python
json_data[0]
```

```python
key = ['cast', 'character', 'genres', 'imdbid', 'label', 'netflixid', 'season', 'title']
```

```python

for i in range(len(json_data)):
    for j in range(len(key)):
        
        try:
            replace = json_data[i][key[j]][0]
        except:
            replace = ''
            json_data[i][key[j]] = replace
            continue
            
        for k in range(1,len(json_data[i][key[j]])):
            replace += json_data[i][key[j]][k]
            
        json_data[i][key[j]] = replace
        del replace
        
    if i==0:
        df_json_netflix = pd.DataFrame(json_data[i], index=[json_data[i]['imdbid']])
    else:
        df_json_netflix = df_json_netflix.append(pd.DataFrame(json_data[i], index=[json_data[i]['imdbid']]))
```

```python
df_json_netflix
```

```python
final_netflix2 = pd.merge(final_netflix, df_json_netflix, left_index=True, right_index=True, how='inner')
final_netflix2
```

```python
final_netflix2.to_csv('final.csv',encoding='utf-8')
```
